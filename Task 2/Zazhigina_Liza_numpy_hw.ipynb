```python
cases = 3000
checks = 500
numpy_time = np.array([])
for number in range (0, cases, 20):
    checks = cases - number
    np_x = np.random.randint(0,100, number + 1)
    np_y = np.random.randint(0,100, number + 1)
    start = time.time()
    for attempt in range (checks):
        np_z = 2*np_x**2+4*np_y
    #numpy_time[0, cases] = (time.time()-start)/5
    numpy_time = np.append(numpy_time, 1000000*(time.time()-start)/checks)
```

```python
def frame(lines, columns, inside, outside):
    entry = np.full((lines - 2, columns - 2), inside)
    hor_frame = np.full((1, columns  - 2), outside)
    ver_frame = np.full((lines, 1), outside)
    entry = np.vstack([entry, hor_frame])
    entry = np.vstack([hor_frame, entry])
    entry = np.hstack([ver_frame, entry])
    entry = np.hstack([entry, ver_frame])
    
    return entry
```

```python
def make_chess(size):
    zero_line = np.zeros(shape =(1,size), dtype = 'int')
    for i in range(size):
        if (i%2 == 1):
                zero_line[0, i] = 1
    
    first_line = np.zeros(shape =(1,size), dtype = 'int')
    for i in range(size):
        if (i%2 == 0):
                first_line[0, i] = 1
                
    m = zero_line
    for i in range(size - 1):
        if (i%2 == 1):
            m = np.vstack([m,zero_line])
        else: m = np.vstack([m,first_line])
    return m
```

```python
def get_board(size): 
    xs = np.arange(0, size[0]) #create an array from 0 to size.first
    ys = np.arange(0, size[1])
    board = np.meshgrid(xs, ys) #creates space, where we can opperate with (x,y)
    return board

def get_init_status(size, initial_prob_life):
    status = np.random.uniform(0, 1, size=size) <= initial_prob_life #uniform creates a probability of concieving life, we compare it to our basic value
    return status #matrix of possible init

def count_live_neighbours(status):
    kernel = np.array([[1,1,1],[1,0,1],[1,1,1]])
    c = convolve2d(status, kernel, mode = 'same', boundary = 'wrap') #конволюция двух матриц
    return c #пространство - тор, соответсвенно, обходит "масочкой" все клетки с подсчетом соседей
def apply_rules(status):
    live_neighbours = count_live_neighbours(status)
    survive_underpop = live_neighbours >= 2
    survive_overpop = live_neighbours <= 3
    survive = status * survive_underpop * survive_overpop
    new_status = np.where(live_neighbours == 3, True, survive) #reproduction
    new_neighbours = count_live_neighbours(new_status)
    return new_status, new_neighbours
    
```

```python
def get_updated_colors(status, c):
    cmap = mpl.cm.YlGn_r
    rescale = c / 8
    colors = [cmap(neighbors) for neighbors in rescale.flatten()]
    is_live = status.flatten()
    colors = [(r, g, b, 0.9) if live else (r, g, b, 0) for live, (r, g, b, a) in zip(is_live, colors)]
    return colors
```

```python
def play_game(size=(50, 50), initial_prob_life=0.5, animation_kwargs=None, marker_size=300):
  
    board = get_board(size)
    status = get_init_status(size, initial_prob_life)
    
    fig, axes = plt.subplots(figsize=(15, 15))
    scatter = axes.scatter(*board, animated=True, s=marker_size, edgecolor=None)
    axes.text(0.5, 0.965, "GAME OF LIFE", transform=axes.transAxes, ha="center", va="bottom", color="lightgreen",
              fontsize=25, family="sans-serif")
    axes.set_facecolor("black")
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)    
        
    def update(frame):
        nonlocal status
        status, live_neigbours = apply_rules(status)
        colors = get_updated_colors(status, live_neigbours)
        scatter.set_facecolor(colors)
        return scatter
    
    animation_kwargs = {} if animation_kwargs is None else animation_kwargs
    return FuncAnimation(fig, update, **animation_kwargs)
```

```python
for i in range(1):
    ani = play_game(
        size=(100, 100),
        marker_size=50,
        initial_prob_life=0.5,
        animation_kwargs={
            "frames": 25,
        }
    )
    ani.save('life_{0}.gif'.format(i), writer='pillow')
```
